```python
import pandas as pd
```

```python
df = pd.read_csv("prolific_survey_1.csv")
df.columns
```

```python
col_english = 'Would you describe yourself as proficient or better in English (able to read, write, and communicate effectively)?'
col_agreement = 'This study has two phases and will last approximately one month. Please indicate whether you agree to participate for the full duration of the study.'
col_medical = 'Have you ever participated in medical training or shadowing in a hospital?'
col_biology = 'Have you completed any advanced (college graduate-level) biology courses?'

df_filtered = df[
    (df[col_english] == 'Yes') &
    (df[col_agreement] == 'Yes') &
    (df[col_medical] == 'No') & 
    (df[col_biology] == 'No')
]

print(f"Total participants initially: {len(df)}")
print(f"Total participants kept (laypeople who agreed): {len(df_filtered)}")
df_filtered.to_csv('filtered_laypeople_study.csv', index=False)
```

```python
df1 = pd.read_csv("removed_more.csv")
print(df1.columns)
df1['What is the highest level of education you have completed?'].value_counts()
df["Participant id"].duplicated().any()
```

```python
edu_col = "What is the highest level of education you have completed?"
approval_col = "Total approvals"
less_than_hs = "Less than High School"
hs = "High school (including GED)"

df_less_than_hs = df1[df1[edu_col] == less_than_hs]
print(df_less_than_hs.shape)

# high school only and take top 60 
df_hs_top60 = (df1[df1[edu_col] == hs].sort_values(by=approval_col, ascending=False)).head(60)
df_degree_top60 = (df1[df1[edu_col].str.contains("degree", case=False, na=False)].sort_values(by=approval_col, ascending=False)).head(60)


df_selected = pd.concat([df_less_than_hs, df_hs_top60, df_degree_top60], ignore_index=True)
```

```python
df_selected.to_csv("survey_2_people.csv", index=False)
```
