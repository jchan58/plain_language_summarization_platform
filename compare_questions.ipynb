```python
import pandas as pd
```

```python
abstracts_df = pd.read_csv(
    r"C:\Users\ercj\Desktop\plain_language_summarization_platform\final_abstracts\0218_combined_abstracts.csv",
    encoding="latin1"
)

final_df = pd.read_csv(
    r"C:\Users\ercj\Desktop\plain_language_summarization_platform\final_user_batches.csv",
    encoding="latin1"
)
```

```python
REPLACE_COLS = [
    "question_1", "question_2", "question_4", "question_5",
    "question_1_answers_choices", "question_1_correct_answers",
    "question_2_answers_choices", "question_2_correct_answers",
    "question_4_answers_choices", "question_4_correct_answers",
    "question_5_answers_choices", "question_5_correct_answers",
]

# build a unique lookup from abstracts_df
abs_key = abstracts_df[["abstract_id"] + REPLACE_COLS].drop_duplicates(subset=["abstract_id"], keep="first")

# merge to pull the "correct" values
tmp = final_df[["abstract_id"]].merge(abs_key, on="abstract_id", how="left")

# sanity: any abstract_id missing in abstracts_df?
missing = tmp[REPLACE_COLS].isna().all(axis=1).sum()
print("Rows where abstract_id not found in abstracts_df:", int(missing))

# apply overwrite (Q3 untouched)
final_df_fixed = final_df.copy()
final_df_fixed[REPLACE_COLS] = tmp[REPLACE_COLS].to_numpy()

# ✅ write a NEW file with header (don't append)
out_path = r"C:\Users\ercj\Desktop\plain_language_summarization_platform\final_user_batches.csv"
final_df_fixed.to_csv(out_path, index=False, encoding="latin1")
print("Wrote:", out_path)
```

```python
final_df = final_df_fixed
```

```python
cols = [
    "question_1","question_2","question_3","question_4","question_5",
    "question_1_answers_choices","question_1_correct_answers",
    "question_2_answers_choices","question_2_correct_answers",
    "question_3_answers_choices","question_3_correct_answers",
    "question_4_answers_choices","question_4_correct_answers",
    "question_5_answers_choices","question_5_correct_answers",
]

subset = final_df[final_df["type"] == "static_2"]

def norm_df(df):
    return df.applymap(lambda x: "" if pd.isna(x) else str(x).strip())

for aid in subset["abstract_id"].unique():
    expected = abstracts_df.loc[abstracts_df["abstract_id"] == aid, ["abstract_id"] + cols].drop_duplicates("abstract_id")
    actual   = subset.loc[subset["abstract_id"] == aid, ["abstract_id"] + cols].drop_duplicates()

    expected_norm = norm_df(expected[cols])
    actual_norm   = norm_df(actual[cols])

    if not expected_norm.equals(actual_norm):
        print("\n==============================")
        print(f"ABSTRACT ID: {aid}")
        print("EXPECTED (from abstracts_df):")
        display(expected)

        print("ACTUAL UNIQUE ROWS (from final_df static_2):")
        display(actual)
```
